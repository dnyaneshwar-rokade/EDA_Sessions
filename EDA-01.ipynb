### Step-03

**Create a Dataframe using lists**

```python
list_01 = [ 'Ram', 'Raheem', 'Robert' ]

list_02 = [ 30, 31, 32 ]

pd.DataFrame ( zip ( list_01, list_02 ) )  # <---- in this output we have not specified any names to columns that's why indexs are coming.
```

```python
list_01 = [ 'Ram', 'Raheem', 'Robert' ]

list_02 = [ 30, 31, 32 ]

pd.DataFrame ( zip ( list_01, list_02 ), columns = [ 'Name', 'Age' ] )  # <---- here we specified column names.
```

```python
list_01 = [ 'Ram', 'Raheem', 'Robert' ]

list_02 = [ 30, 31, 32 ]

col_names = [ 'Name', 'Age' ]   # <------  we can also specify column name in a variable and later use it. like below we used it.

names = pd.DataFrame ( zip ( list_01, list_02 ), columns = col_names )  # <----- we saved this dataframe in a variable to print it later.
```

**How to drop columns/rows**

```python
# what is your data frame name
# i want to drop column
************************ Axix **************************
# axis = 1   ========> means it is about columns.
# axis = 0   ========> means it is about rows.
********************************************************
# what is the name of the column ? : City
********************** inplace *************************
# when you are droping the column means
# you are modifying the table
# the dataframe should be overwrite or 
# you want to create a new dataframe  

# inplace=True  =====> means table will overwrite 
```
