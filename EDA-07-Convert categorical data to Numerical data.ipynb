$Method-2:np.where()$

```python
file_path = "C:\\Users\\DBRok\\Downloads\\Visadataset.csv"
visa_df = pd.read_csv(file_path)
visa_df.head()
```

- np.where() is applicable for binary conditions

- which means it is applicable only for two lables

- np.where(condition,Truevalue,False_value)

- For example case_status has two lables

- condition:  <col_values>=='Certified'
    
- True value: Replace all certified values with 0
    
- False value: Replace all denied values with 1

```python
con = visa_df['case_status']=='Certified'

visa_df['case_status']=np.where(con,0,1)
```

```python
############# Read the data again ############
file_path = "C:\\Users\\DBRok\\Downloads\\Visadataset.csv"
visa_df = pd.read_csv(file_path)
print(visa_df[['continent','case_status']].head(10))

# step-01
from sklearn.preprocessing import LabelEncoder

# step-02
le=LabelEncoder()

# step-03
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df['continent']=le.fit_transform(visa_df['continent'])
print(visa_df[['continent','case_status']].head(10))
```

**One hot encoder**

- one hot encoder means at a time only one will be ON(1/True), others are OFF(0/False)

- Supposer case status has two unique lables

- Certified and denied

- In one hot encoder it will create two new columns

- like below

|case_status|certified|denied|
|-----------|---------|------|
|certified|1|0|
|denied|0|1|


**Advantage**

- One hot encoder new columns are orthogonal each other

- Orthogonality means 90 degress phase shift

- 90 degress phase shift means perpendicular each other

- Perpendicular means independent each other

- Independent means no relation

- It is very important property that is to avoid relation between variabels

**Draw back**:

- Assume that you have 100 unique lables for a column

- So it will create 100 new columns

- To process 100 columns we require more memory,more time,more hardaware 

- This is called **Curse of Dimensionality**
